# Figures for psKWR results


This compares performance of models trained on the joint probabilities P(m,n) using Hellinger distance and those trained to minimize P(m | n) using truncated KL divergence.


Comparisons include KL divergence, Mean Squared Error, and Hellinger distance on testing parameters while:

1. varying number of kernel density functions,
2. varying number of nodes,
3. varying number of hidden layers, and
4. varying number of training parameters.

It finally shows how we selection of the best model based on validation data.

```python
def plot_metrics(hell_mean_array_,
                 hell_std_array_,
                 kld_mean_array_,
                 kld_std_array_,
                 mse_mean_array_,
                 mse_std_array_,
                 x,
                 x_label,
                 colors,
                 fs = 23,
                 alpha = 0.1,
                save_path = None,
                line_color='k',
                tick_param_size = 22):

    fig,ax = plt.subplots(1,3,figsize=(18,5))
    n=0
    ax[0].plot(x,hell_mean_array_[:],color=line_color)
    ax[0].scatter(x,hell_mean_array_[:],s=20,color=line_color)
    ax[0].fill_between(x,hell_mean_array_[:]-hell_std_array_[:],hell_mean_array_[:]+hell_std_array_[:],
                      alpha=alpha,color=colors[n])
    ax[0].set_ylabel('Hellinger distance',fontsize=fs)
    
    ax[1].plot(x,kld_mean_array_[:],color=line_color)
    ax[1].scatter(x,kld_mean_array_[:],s=20,color=line_color)
    ax[1].fill_between(x,kld_mean_array_[:]-kld_std_array_[:],kld_mean_array_[:]+kld_std_array_[:],
                      alpha=alpha,color=colors[n])
    ax[1].set_ylabel('KL divergence',fontsize=fs)
    
    ax[2].plot(x,mse_mean_array_[:],color=line_color)
    ax[2].scatter(x,mse_mean_array_[:],s=20,color=line_color)
    ax[2].fill_between(x,mse_mean_array_[:]-mse_std_array_[:],mse_mean_array_[:]+mse_std_array_[:],
                      alpha=alpha,color=colors[n])
    ax[2].set_ylabel('Mean squared error',fontsize=fs)


    for i in range(3):
        ax[i].tick_params(axis='both',labelsize=tick_param_size)
        ax[i].set_xticks(x,labels=x)
        ax[i].set_yscale('log')
        ax[i].set_xlabel(x_label,fontsize=fs)

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path,dpi=450,bbox_inches='tight');
```

## Figure S4 a-d: KLD trained model performance on testing parameters

```python
# load in testing parameters
npdf = 10
num_files = 2
test_ds = tm2.burstyBernoulliDataset([f'../data/psKWR_data/{npdf}_npdf_cheb/256_params_{i}_test'
                                      for i in range(num_files)])

params_10 = test_ds.get_data_dict(None)['params']
cond_pss_10 = test_ds.get_data_dict(None)['cond_pss']

npdf = 5
test_ds = tm2.burstyBernoulliDataset([f'../data/psKWR_data/{npdf}_npdf_cheb/256_params_{i}_test'
                                      for i in range(num_files)])

params_5 = test_ds.get_data_dict(None)['params']
cond_pss_5 = test_ds.get_data_dict(None)['cond_pss']

npdf = 3
test_ds = tm2.burstyBernoulliDataset([f'../data/psKWR_data/{npdf}_npdf_cheb/256_params_{i}_test'
                                      for i in range(num_files)])

params_3 = test_ds.get_data_dict(None)['params']
cond_pss_3 = test_ds.get_data_dict(None)['cond_pss']


# SAME number of params/cond_pss each
N = min(len(params_3),len(params_5),len(params_10))
params_10 = params_10[:N,:]
params_5 = params_5[:N,:]
params_3 = params_3[:N,:]

cond_pss_10 = cond_pss_10[:N,:]
cond_pss_5 = cond_pss_5[:N,:]
cond_pss_3 = cond_pss_3[:N,:]
```

```python
# calculate testing metrics
eps = 1e-18
kld_metric_dict = {p : {} for p in kld_paths}
kld_metric_dict = {p : {} for p in kld_paths}
kld_metric_dict = {p : {} for p in kld_paths}

# over grid of length 300
m_limit = 300
m = torch.arange(m_limit)   

for i,path in enumerate(kld_paths):
    print(i,path)
    if '10npdf' in path:
        params = params_10
        cond = cond_pss_10.detach().numpy() 
        n_vals = params_10[:,3]
        b,beta,gamma = 10**params_10[:,0],10**params_10[:,1],10**params_10[:,2]
        mu_n = b/beta
        var_n = (mu_n)*(b + 1)
        n_binom = mu_n**2/(var_n - mu_n)
        p_binom = mu_n/var_n
        # nascent marginal probability
        comb =  torch.lgamma(n_vals + n_binom) - torch.lgamma(n_binom) - torch.lgamma(n_vals + 1) 
        log_p_nascent = comb + n_binom*torch.log(p_binom) + n_vals*torch.log(1-p_binom)

    elif '5npdf' in path:
        params = params_5
        cond = cond_pss_5.detach().numpy()
        n_vals = params_5[:,3]
        b,beta,gamma = 10**params_5[:,0],10**params_5[:,1],10**params_5[:,2]
        mu_n = b/beta
        var_n = (mu_n)*(b + 1)
        n_binom = mu_n**2/(var_n - mu_n)
        p_binom = mu_n/var_n
        # nascent marginal probability
        comb =  torch.lgamma(n_vals + n_binom) - torch.lgamma(n_binom) - torch.lgamma(n_vals + 1) 
        log_p_nascent = comb + n_binom*torch.log(p_binom) + n_vals*torch.log(1-p_binom)

    elif '3npdf' in path:
        params = params_3
        cond = cond_pss_3.detach().numpy()
        n_vals = params_5[:,3]
        b,beta,gamma = 10**params_5[:,0],10**params_5[:,1],10**params_5[:,2]
        mu_n = b/beta
        var_n = (mu_n)*(b + 1)
        n_binom = mu_n**2/(var_n - mu_n)
        p_binom = mu_n/var_n
        # nascent marginal probability
        comb =  torch.lgamma(n_vals + n_binom) - torch.lgamma(n_binom) - torch.lgamma(n_vals + 1) 
        log_p_nascent = comb + n_binom*torch.log(p_binom) + n_vals*torch.log(1-p_binom)
        
    model = torch.load(path+'/MODEL')
    # predicted
    pred = model.get_cond_P(params, m_limit)
    pred = torch.exp(log_p_nascent[:,None])*pred
    pred = pred.detach().numpy()
    pred[pred<eps] += (eps - pred[pred<eps])
    cond[cond<eps] += (eps - cond[cond<eps])
    
    kld_metric_dict[path]['hell'] = np.sum( (np.sqrt(cond) - np.sqrt(pred))**2, axis = 1)/2
    kld_metric_dict[path]['mse'] = np.sum((cond-pred)**2, axis = 1)
    
    # normalize before calculating KLD
    cond = cond / (cond.sum(axis=1)[:,None])
    pred = pred / (pred.sum(axis=1)[:,None])
    kld_metric_dict[path]['kld'] = np.sum(cond * np.log(cond/pred), axis = 1)
```
