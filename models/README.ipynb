## Models




This folder contains the final architecture for kernel weight regression (KWR) (best_model_MODEL) and its metadata (best_model_meta.npy). It also contains the following folders: 



----

### kernel_models: Varying Number of Kernel Functions

Contains models and metadata with architectures that produce weights for 1, 5, 7, and 20 kernel functions (and single scaling factor):


| Name      | Input dim | Layer 1 dim | Layer 2 dim | Output dim (weights)| Output dim (scaling factor) | # Training Examples (Conditional) | # Training Examples (Rate Vectors) |
| ----------- | ----------- | ----------- | ----------- | ----------- | ----------- |----------- | ----------- |
| 1npdf_MODEL     | 7       | 256       | 256       | 1       | 1       | 211,515 | 1536 | 
| 5npdf_MODEL     | 7       | 256       | 256       | 5       | 1       | 211,515 | 1536 | 
| 7npdf_MODEL     | 7       | 256       | 256       | 7       | 1       | 211,515 | 1536 | 
| 15npdf_MODEL     | 7       | 256       | 256       | 15       | 1       | 211,515 | 1536 | 
| 20npdf_MODEL     | 7       | 256       | 256       | 20       | 1       | 211,515 | 1536 |

### direct_models:  Architectures for direct regression

Contains models for directly predicting probability of single microstate (direct regression), architectures with various number of hidden layers and nodes per hidden layer. Naming convention: 


| Name | Input dim |  # Nodes per hidden layer | # Hidden layers |Output dim |
| ----------- | ----------- | ----------- | ----------- | ----------- |
| **X**hidu_**Y**hidl_MODEL   | 5       | **X**       | **Y**       | 1       |
